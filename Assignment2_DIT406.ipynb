# --------------------------------------------------------------------------------------------------------------
## Question 1. b)
#### What are the values of the slope and intercept of the regression line?
# --------------------------------------------------------------------------------------------------------------

```python
# Print some interesting metrics.
print('R-value:', model.score(x, y))
print('Intercept:', model.intercept_)
print('Slope:', model.coef_[0])
print('Minimum Living area', min(x)[0])
print('Maximum Living area', max(x)[0])
```

The simple linear regression we use in 1. a) looks like this: $ E[Y] = \beta_0 + \beta_1 \cdot X $.

The R-value is how much of the data we are able to explain using our model. In this case the R-value is $0.31579478122912397 \approx 31.579 \%$ which means that we are able to explain $31.579\%$ of the data with our model.

The intercept is what value we are left with when $X=0$ i.e $E[Y]=\beta_0$

So $\beta_0 = 2220603.243355869:-$ and means something like, if there would be a an object with zero Living area it would cost
around 2220603 :- which is pretty absurd since one simply cannot buy/sell an object without living area.

Note that data we received from Hemnet is "just" objects with [63, 210] in living area. Thus interpolation (extrapolation) outside this interval is in some sense dangerous as we know nothing outside of this interval.


The slope of the model is in this case $\beta_1 = 19370.13854733158$ which basicly means that for each additional $m^2$ we expect the price to go up with this value of $\beta_1$.

## Question 2. b)
#### Use k-nearest neighbours to classify the iris data set with some different values for k, and with uniform and distance-based weights. What will happen when k grows larger for the different cases? Why?
# --------------------------------------------------------------------------------------------------------------

```python
# max number of k in KNN.
max_k = 90

# Create two different arrays to store accuracy values.
acc_uniform = np.zeros(max_k)
acc_distance = np.zeros(max_k)


for k in range(0, max_k):
    # fit the KNN model using uniform
    knn_uniform = KNN(n_neighbors = k+1, weights = 'uniform')
    knn_uniform.fit(x_train,y_train)
    knn_pred_uniform = knn_uniform.predict(x_test)
    
    # fit the KNN model using distance
    knn_distance = KNN(n_neighbors = k+1, weights = 'distance')
    knn_distance.fit(x_train,y_train)
    knn_pred_distance = knn_distance.predict(x_test)
    
    
    # Calculating metrics
    acc_uniform[k] = metrics.accuracy_score(y_test, knn_pred_uniform)
    acc_distance[k] = metrics.accuracy_score(y_test, knn_pred_distance)
```

```python
# Plot Number of K-neighbours on the x-axis and Accuracy on y-axis for both of the two methods Uniform and Distance.
plt.gcf().set_size_inches(12, 8)
plt.plot(np.linspace(1, max_k, max_k), acc_uniform[:],label = 'Weights = Uniform')
plt.plot(np.linspace(1, max_k, max_k), acc_distance[:], label = 'Weights = Distance')
plt.legend()
plt.xlabel('K-neighbours')
plt.ylabel('Accuracy')
plt.title('K-neighbours vs Accuracy')
plt.show()
```

From the plot above one can note that when we use uniform accuracy decreases dramatically around 60 number of neighbours i.e k = 60. Whereas when we use distance the accuracy is stable for all different values of k.
