**Question 1.**
The branching factor ‘d’ of a directed graph is the maximum number of children (outer degree) of
a node in the graph. Suppose that the shortest path between the initial state and a goal is of length ‘r’.

**a)**
What is the maximum number of Breadth First Search (BFS) iterations required to reach the
solution in terms of ‘d’ and ‘r’?

**b)**
Suppose that storing each node requires one unit of memory and the search algorithm stores
each entire path as a string of nodes. Hence, storing a path with k nodes requires k units of
memory. What is the maximum amount of memory required for BFS in terms of ‘d’ and ‘r’ ?

**Answer to Question 1. a):**
Okay, so below is a example of a directed graph with 4 Different levels, the branching factor is d = 2 and the shortest path to the goal is r = 3. The different coloring shows how the BFS algorithm works in each iteration.

Thus we can create a general formula as following:
$$ Max_{iter} = \sum _ {l = 0} ^{r-1} d^l = \frac{d^r}{d-1} - \frac {d}{d-1} = \frac {d^r-d}{d-1}$$


In this Example below the maximum BFS iteration is 7 and can be calculated as: 
$$ Max_{iter} = \sum _ {l = 0} ^ {3-1} 2^l = 2^0+2^1+2^2 = 1+2+4=7 $$
